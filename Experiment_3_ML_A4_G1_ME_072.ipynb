```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_wine = pd.read_csv('winequalityN.csv')
df_wine.head()
```

```python
df_wine.describe()
```

```python
df_wine.info()
```

```python
# check for null values
df_wine.isnull().sum()
```

```python
# fill the missing values using mean value statistics obtained using mean() function
for col, value in df_wine.items():
    if col != 'type':
        df_wine[col] = df_wine[col].fillna(df_wine[col].mean())
```

```python
df_wine.isnull().sum()
```

```python
# create box plots
fig, ax = plt.subplots(ncols=6, nrows=2, figsize=(20,10))
index = 0
ax = ax.flatten()

for col, value in df_wine.items():
    if col != 'type':
        sns.boxplot(y=col, data=df_wine, ax=ax[index])
        index += 1
plt.tight_layout(pad=0.5, w_pad=0.7, h_pad=5.0)
```

```python
# create dist plot
fig, ax = plt.subplots(ncols=6, nrows=2, figsize=(20,10))
index = 0
ax = ax.flatten()

for col, value in df_wine.items():
    if col != 'type':
        sns.histplot(value, ax=ax[index])
        index += 1
plt.tight_layout(pad=0.5, w_pad=0.7, h_pad=5.0)
```
